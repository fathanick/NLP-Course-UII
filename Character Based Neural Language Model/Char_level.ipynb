###Load and View Data

```
def load_file(filename):
  f = open(filename,'r')
  txt = f.read()
  f.close()

  return txt
```

```
myfile = "/content/drive/My Drive/PYTHON CODE OF MINE/sample-text.txt"
my_text = load_file(myfile)
print(my_text)
```

###Clean Text

```
tokens = my_text.split()
my_text = ' '.join(tokens)
my_text
```

###Create Sequences

```
length = 10
seq = list()

for i in range(length, len(my_text)):
  sq = my_text[i-length:i+1]
  seq.append(sq)
print("Total sequences: %d" % len(seq))
```

###Save Sequences

```
def save_file(lines, filename):
  data = '\n'.join(lines)
  file = open(filename,'w')
  file.write(data)
  file.close()
```

```
output_file = '/content/drive/My Drive/PYTHON CODE OF MINE/char_seq.txt'
save_file(seq, output_file)
```

###Train Language Model


####Load Data

```
char_file = '/content/drive/My Drive/PYTHON CODE OF MINE/char_seq.txt'
raw_txt = load_file(char_file)
lines = raw_txt.split('\n')
#print(lines)
```

####Encode Sequences

```
chars = sorted(list(set(raw_txt)))
mapping = dict((c,i) for i, c in enumerate(chars))
#print(chars)
#print(mapping)
```

```
seq = list()

for line in lines:
  #integer encode line
  encoded_seq = [mapping[char] for char in line]
  #store
  seq.append(encoded_seq)

print(seq)
```

```
#vocab size
vocab_size = len(mapping)
print("Vocabulary size: %d" % vocab_size)
```

####Split Inputs and Output

```
from numpy import array

seq = array(seq)
X,y = seq[:,:-1], seq[:,-1]
```

####One Hot Encode

```
from keras.utils import to_categorical

seq = [to_categorical(x, num_classes = vocab_size) for x in X]
X = array(seq)
y = to_categorical(y, num_classes = vocab_size)
print(X)
print(y)
```

####Fit Model

```
from keras.models import Sequential
from keras.layers import Dense, LSTM
#define model

model = Sequential()
model.add(LSTM(64, input_shape=(X.shape[1], X.shape[2])))
model.add(Dense(vocab_size, activation = 'softmax'))
print(model.summary())
```

```
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=100, verbose=2)
```

####Save Model

```
from pickle import dump
#save the model
model.save('/content/drive/My Drive/PYTHON CODE OF MINE/model.h5')

#save the mapping
dump(mapping, open('/content/drive/My Drive/PYTHON CODE OF MINE/mapping.pkl','wb'))
```

```
from pickle import load
from keras.models import load_model
from keras.utils import to_categorical
from keras.preprocessing.sequence import pad_sequences

# generate a sequence of characters with a language model
def generate_seq(model, mapping, seq_length, seed_text, n_chars):
	in_text = seed_text
	# generate a fixed number of characters
	for _ in range(n_chars):
		# encode the characters as integers
		encoded = [mapping[char] for char in in_text]
		# truncate sequences to a fixed length
		encoded = pad_sequences([encoded], maxlen=seq_length, truncating='pre')
		# one hot encode
		encoded = to_categorical(encoded, num_classes=len(mapping))
		#print(encoded.shape[1])
		#encoded = encoded.reshape(1, encoded.shape[0], encoded.shape[1])
		# predict character
		yhat = model.predict_classes(encoded, verbose=0)
		# reverse map integer to character
		out_char = ''
		for char, index in mapping.items():
			if index == yhat:
				out_char = char
				break
		# append to input
		in_text += char
	return in_text

model = load_model('/content/drive/My Drive/PYTHON CODE OF MINE/model.h5')
mapping = load(open('/content/drive/My Drive/PYTHON CODE OF MINE/mapping.pkl','rb'))

# test start of rhyme
print(generate_seq(model, mapping, 10, 'bintang raksa', 20))
# test mid-line
print(generate_seq(model, mapping, 10, 'kenal dengan na', 20))
# test not in original
print(generate_seq(model, mapping, 10, 'bintang Supernov', 20))
```
