```python
#COUNT WORD FREQUENCY

from collections import Counter

word_dict = Counter(" ".join(df['text'].values.tolist()).split(" ")).items()
print(word_dict)
```

```python
df_word = pd.DataFrame(word_dict, columns=['word','frequency'])
df_word.head()
```

```python
#sort by frequency
sorted_word = df_word.sort_values(by=['frequency'], ascending=False).head()
print(sorted_word)
```

```python
#show vertical bar
fig, ax = plt.subplots(figsize=(8,8))

sorted_word.plot.bar(x='word', 
                      y='frequency',
                      ax=ax,
                      color='blue')

ax.set_title("Most common words")
plt.show()
```

```python
#show horizontal bar
fig, ax = plt.subplots(figsize=(8,8))

sorted_word.plot.barh(x='word', 
                      y='frequency',
                      ax=ax,
                      color='blue')

ax.set_title("Most common words")
plt.show()
```

```python
#simple way
#read this reference: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.Series.plot.html

sorted_word.plot(kind='bar', x='word', color="blue")
plt.xlabel("Word")
plt.ylabel("Word Frequency")
plt.title("Most Common Word")
plt.show()

#save image
plt.savefig("bar-plot.png", format="png")
```

```python
import nltk
from nltk.tokenize import word_tokenize
```

```python
#VISUALIZING N-GRAMS 
# https://towardsdatascience.com/from-dataframe-to-n-grams-e34e29df3460

#create list of words from text column in dataframe
sent_list = df.text.tolist()
word_list = []
for s in sent_list:
    w = word_tokenize(s)
    word_list.append(w)

print(word_list) 
```

```python
#the word_list is in a form list of lists. In this case, we need to flatten the word_list 
words = sum(word_list, [])

print(words[:10])
```

```python
bigram_series = (pd.Series(nltk.ngrams(words, 2)).value_counts())[:10]
bigram_series
```

```python
bigram_series.sort_values().plot.barh(color='blue', width=.9, figsize=(12, 8))
plt.title('20 Most Frequently Occuring Bigrams')
plt.ylabel('Bigram')
plt.xlabel('# of Occurances')
```
